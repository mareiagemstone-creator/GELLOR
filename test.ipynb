```python
lib = ["numpy", "pandas", "matplotlib", "scipy", "stats", "statsmodels"]

for pkg in lib:
    try:
        # We replace "stats" with "statsmodels" for a more accurate check,
        # as "stats" isn't a common top-level library to import directly.
        if pkg == "statsmodels":
            import statsmodels.api as sm
        elif pkg == "stats":
            # This package usually isn't a direct pip install name, it might be a module
            # within scipy or another library, so we print a note instead.
            print(f"'{pkg}' isn't a standard top-level library name to import directly.")
            continue
        else:
            __import__(pkg) # This is a generic way to import by string name

        print(f"✅ FOUND: The '{pkg}' library is installed.")

    except ImportError:
        print(f"❌ MISSING: The '{pkg}' library is NOT installed.")
        print(f"   You can install it with: pip install {pkg}")
    except Exception as e:
        print(f"⚠️ An error occurred while checking {pkg}: {e}")
    print("-" * 20)

  
```

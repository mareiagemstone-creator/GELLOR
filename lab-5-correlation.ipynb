```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'C:\Users\mariagemstone\Downloads\current-power.csv')
df
```

```python
df. describe()
```

```python
df. cov()
```

```python
x=df['Current']
y=df[ 'Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power')
plt.xlabel('Current')
plt.show()
```

```python
df. corr ()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(10,5))
plt.scatter(
    x,y,
    color = '#1f77b4',
    alpha=0.6,
    label='Data points'     
)  

plt.plot(
    x,
    line,
    color = '#ff7f0e',
    label=f'Regression line: (r= {r_value:.3f})'
)
plt. title('Device Current vs Power Response with Regression Line')
plt. ylabel('Power')
plt. xlabel('Current')
plt. legend()
plt. show()

```

## AIR QUALITY <hr style="border:2.5px solid rgba(221, 83, 19, 0.82)"></hr>

```python
air = pd.read_csv(r'C:\Users\mariagemstone\Downloads\air-quality.csv')
air
```

## NOx_Humidity <hr style="border:2.5px solid rgba(221, 83, 19, 0.82)"></hr>

```python
x=air['NOx']
y=air[ 'Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(10,5))
plt.scatter(
    x,y,
    color = '#1f77b4',
    alpha=0.6,
    label='Data points'     
)  

plt.plot(
    x,
    line,
    color = '#ff7f0e',
    label=f'Regression line: (r= {r_value:.3f})'
)
plt. title('NOx vs Humidity Correlation')
plt. ylabel('Relative Humidity')
plt. xlabel('NOx (g/m3)')
plt. legend()
plt. show()

```

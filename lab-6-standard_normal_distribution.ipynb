## Histogram <hr style="border:2.5px solid rgba(232, 13, 181, 0.59)"></hr>

```python
plt.hist(df['Current'],
         bins=5,
         alpha=0.8,
         color = 'Blue',
         edgecolor = 'Black')

plt.title('Histogram of Current Values',
         font = 'Palatino Linotype',
         fontsize= 12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

## Noramlization <hr style="border:2.5px solid rgba(232, 13, 181, 0.59)"></hr>

```python
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x,p,
         color = 'Blue',
         linewidth = 2)

plt.title('Noraml Distribution of Current Values',
         font = 'Palatino Linotype',
         fontsize= 12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()
```

```python
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x,p,
         color = 'Blue',
         linewidth = 2)

plt.hist(df['Current'],
         bins=5,
         density= True,
         alpha=0.8,
         color = 'Orange',
         edgecolor = 'Black')


plt.title('Noraml Distribution of Current Values',
         font = 'Palatino Linotype',
         fontsize= 12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()

```

## Standard Normal Distribution N(0,1) <hr style="border:2.5px solid rgba(232, 13, 181, 0.59)"></hr>

```python
mean = df['Z-score'].mean()
std = df['Z-score'].std(ddof=1)

x_min = df['Z-score'].min()
x_max = df['Z-score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)



plt.plot(x,p,
         color = 'Blue',
         linewidth = 2)


plt.title('Standard Normal Distribution N(0,1)',
         font = 'Palatino Linotype',
         fontsize= 12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()

```

```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r"C:\Users\mariagemstone\OneDrive\Desktop\GELLOR\Datasets\transformer-voltage.csv")
df   
```

```python
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)
z_critical
```

```python
pop_std = 5

ci_lower, ci_upper = stats.norm.interval(
confidence = 0.95,  
loc = df['Voltage'].mean(),
scale = pop_std/np.sqrt(df['Voltage'].count())
)

df['Voltage'].mean(), ci_lower, ci_upper
print(f"CI: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

x_min = df['Voltage'].min() 
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x,p,
         color = 'Black',
         linewidth = 2)

plt.hist(df['Voltage'],
         bins=5,
         density= True,
         alpha=0.8,
         color = 'Pink',
         edgecolor = 'Black')


plt.title('Voltage Distribution',
         font = 'Palatino Linotype',
         fontsize= 12)
plt.xlabel('Voltage (V)')
plt.ylabel('Density')
plt.show()

```
